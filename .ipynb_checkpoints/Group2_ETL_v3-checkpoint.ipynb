## TRANSFORM DataFrames

### TRANSFORM DataFrame Gun Violence

```python
# Check gun_df columns
for col in gun_df.columns:
    print(col)
```

```python
# Select needed columns
gun_cols = ["incident_id",
            "date",
            "state",
            "address",
            "n_killed",
            "n_injured",
            "latitude",
            "longitude"]
            
gun_cleaned_df= gun_df[gun_cols].copy()
gun_cleaned_df
```

```python
# Filter to Florida incidents

gun_FL_df = gun_cleaned_df[gun_cleaned_df['state'] == "Florida"]
gun_FL_df
```

```python
gun_FL_df = gun_FL_df.dropna()
gun_FL_df
```

```python
# create list empty for cities
cities = []

# loop on gun florida dataframe
for col in gun_FL_df.columns:
    lats = gun_FL_df['latitude']
    lngs = gun_FL_df['longitude']    
# create a tuple with both latitudes and longitudes
    lat_lngs = zip(lats, lngs)
    
#     Identify nearest city for each lat, lng combination, use citipy,
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
#     # If the city is not already in our list, then add it to cities list
#     if city not in cities:
    cities.append(city)
cities
```

```python
# Print the city count to confirm sufficient count
len(cities)
```

```python
gun_FL_df['city'] = cities
gun_FL_df
```

```python
# Reset index of dataframe
gun_FL_df = gun_FL_df.reset_index(drop=True)
gun_FL_df
```

### TRANSFORM DataFrame Income

```python
for col in income_df.columns:
    print(col)
```

```python
# Select needed columns
income_cols = ["State_Name",
            "State_ab",
            "County",
            "City",
            "Mean",
            "Median"]
            
income_cleaned_df= income_df[income_cols].copy()
income_cleaned_df
```

```python
income_cleaned_df = income_cleaned_df.rename(columns={"State_Name": 'state',
                                                        "City": "city",
                                                      "Mean": "mean",
                                                      "Median": "median"
                                                       })
income_cleaned_df
```

```python
# Florida income of cities and counties filtered

income_FL_df = income_cleaned_df[income_cleaned_df['State_ab'] == "FL"]
income_FL_df
```

```python
income_FL_df = income_FL_df.drop_duplicates("city")
income_FL_df
```

```python
# city name to lowercase to avoid problems merging
income_FL_df["city"] = income_FL_df["city"].str.lower()
income_FL_df
```

```python
income_FL_df = income_FL_df.reset_index(drop=True)
income_FL_df
```

```python
gun_FL_merged_1_df = pd.merge(gun_FL_df, income_FL_df, how='left', on='city')
gun_FL_merged_1_df
```

```python
# DRop empty fields after merging
gun_FL_merged_1_df = gun_FL_merged_1_df.dropna()
gun_FL_merged_1_df
```

```python
# drop duplicated state column after merged
gun_FL_merged_1_df = gun_FL_merged_1_df.drop(columns=['state_y'])
gun_FL_merged_1_df
```

```python
# rename state column after merged
gun_FL_merged_1_df = gun_FL_merged_1_df.rename(columns={"state_x": "state"})
```

```python
# reset index for merged dataframe
gun_FL_merged_1_df = gun_FL_merged_1_df.reset_index(drop=True)
gun_FL_merged_1_df
```

```python
# Create a filtered dataframe from specific columns
premise_cols = ["License Serial Number", "Premises Name", "County ID Code"]
premise_transformed= premise_df[premise_cols].copy()

# Rename the column headers
premise_transformed = premise_transformed.rename(columns={"License Serial Number": "id",
                                                          "Premises Name": "premise_name",
                                                          "County ID Code": "county_id"})

# Clean the data by dropping duplicates and setting the index
premise_transformed.drop_duplicates("id", inplace=True)
premise_transformed.set_index("id", inplace=True)

premise_transformed.head()
```

### Transform county DataFrame

```python
county_cols = ["ID", "County Name (Licensee)", "County ID Code", "License Count"]
county_transformed = county_df[county_cols].copy()

# Rename the column headers
county_transformed = county_transformed.rename(columns={"ID": "id",
                                                         "County Name (Licensee)": "county_name",
                                                         "License Count": "license_count",
                                                         "County ID Code": "county_id"})

# Set index
county_transformed.set_index("id", inplace=True)

county_transformed.head()
```

### Create database connection

```python
connection_string = "postgres:postgres@localhost:5432/customer_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
premise_transformed.to_sql(name='premise', con=engine, if_exists='append', index=True)
```

```python
county_transformed.to_sql(name='county', con=engine, if_exists='append', index=True)
```
