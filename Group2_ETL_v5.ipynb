## TRANSFORM DataFrames

### TRANSFORM DataFrame Gun Violence

```python
# Check gun_df columns
for col in gun_df.columns:
    print(col)
```

```python
# Select needed columns
gun_cols = ["incident_id",
            "date",
            "state",
            "address",
            "n_killed",
            "n_injured",
            "latitude",
            "longitude"]
            
gun_cleaned_df= gun_df[gun_cols].copy()
gun_cleaned_df.head()
```

```python
# Filter to Florida incidents

gun_FL_df = gun_cleaned_df[gun_cleaned_df['state'] == "Florida"]
gun_FL_df.head()
```

```python
# drop NaN values
gun_FL_df = gun_FL_df.dropna()
gun_FL_df.head()
```

```python
# Geolocate cities / towns of gun incidents using latitude and longitude with Citipy
# create list empty for cities
cities = []

# loop on gun florida dataframe
for col in gun_FL_df.columns:
    lats = gun_FL_df['latitude']
    lngs = gun_FL_df['longitude']    
# create a tuple with both latitudes and longitudes
    lat_lngs = zip(lats, lngs)
    
#     Identify nearest city for each lat, lng combination, use citipy,
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
#     # If the city is not already in our list, then add it to cities list
#     if city not in cities:
    cities.append(city)
cities
```

```python
# Print the city count to confirm sufficient count
len(cities)
```

```python
gun_FL_df['city'] = cities
gun_FL_df
```

```python
# Reset index of dataframe
gun_FL_df = gun_FL_df.reset_index(drop=True)
gun_FL_df.head()
```

### TRANSFORM DataFrame Income

```python
for col in income_df.columns:
    print(col)
```

```python
# Select needed columns only for a dataframe
income_cols = ["State_Name",
            "State_ab",
            "County",
            "City",
            "Mean",
            "Median"]
            
income_cleaned_df= income_df[income_cols].copy()
income_cleaned_df.head()
```

```python
# Rename some columns to avoid further problems in merge
income_cleaned_df = income_cleaned_df.rename(columns={"State_Name": 'state_name',
                                                        "City": "city",
                                                      "Mean": "mean_income",
                                                      "Median": "median_income"
                                                       })
income_cleaned_df.head()
```

```python
# Florida income of cities and counties filtered

income_FL_df = income_cleaned_df[income_cleaned_df['State_ab'] == "FL"]
income_FL_df.head()
```

```python
income_FL_df = income_FL_df.drop_duplicates("city")
income_FL_df.head()
```

```python
# city name to lowercase to avoid problems merging
income_FL_df["city"] = income_FL_df["city"].str.lower()
income_FL_df.head()
```

```python
# reset index
income_FL_df = income_FL_df.reset_index(drop=True)
income_FL_df.head()
```

```python
gun_FL_merged_1_df = pd.merge(gun_FL_df, income_FL_df, how='left', on='city')
gun_FL_merged_1_df.head()
```

```python
# Drop empty fields after merging
gun_FL_merged_1_df = gun_FL_merged_1_df.dropna()
# gun_FL_merged_1_df
```

```python
# drop duplicated state column after merged
gun_FL_merged_1_df = gun_FL_merged_1_df.drop(columns=['state'])
gun_FL_merged_1_df.head()
```

```python
# rename state column after merged
gun_FL_merged_1_df = gun_FL_merged_1_df.rename(columns={"State_ab": "state_ab",
                                                        "County": "county",
                                                       })
```

```python
# reset index for merged dataframe
gun_FL_merged_1_df = gun_FL_merged_1_df.reset_index(drop=True)
gun_FL_merged_1_df.head()
```
